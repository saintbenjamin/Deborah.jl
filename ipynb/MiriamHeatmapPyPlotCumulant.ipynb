```julia
const PROJECT_ROOT = joinpath(homedir(), "work", "Deborah.jl")
function localpath(relpath::AbstractString)
    return joinpath(PROJECT_ROOT, relpath)
end
# using Revise
import Deborah.Rebekah; flush(stdout); flush(stderr)
import Deborah.RebekahMiriam;  flush(stdout); flush(stderr)
```

```julia
SAVE_DIR="sample"
ns="12"
nt="4"
beta="1.60"
# model="Plaq-Rect-Loop-TrM1_GBM_Plaq-Rect-Loop-TrM2_GBM_Plaq-Rect-Loop-TrM3_GBM_Plaq-Rect-Loop-TrM4_GBM"
model="Plaq-Rect-TrM1_GBM_Plaq-Rect-TrM2_GBM_Plaq-Rect-TrM3_GBM_Plaq-Rect-TrM4_GBM"
# model="TrM1-TrM1_BAS_TrM1-TrM2_GBM_TrM1-TrM3_GBM_TrM1-TrM4_GBM"
```

```julia
overall_name="L$(ns)T$(nt)b$(beta)_$(model)"

# --- Load result ---
result = Rebekah.JLD2Loader.load_jld2_Miriam(localpath("$(SAVE_DIR)/results_$(overall_name).jld2"))

rw_data = result.rw_data
new_dict = result.summary
new_dict_trace_meas = result.summary_trace_meas
new_dict_moment_meas = result.summary_moment_meas
new_dict_cumulant_meas = result.summary_cumulant_meas
kappa_list = result.kappa_list
labels = result.labels
trains = result.trains

labels_int = parse.(Int, labels)
trains_int = parse.(Int, trains)   

Rebekah.PyPlotLaTeX.set_pyplot_latex_style()

save_flag=true
```

```julia
Y_tags = [:Y_BS, :Y_JK, :Y_P1, :Y_P2]
first_block_keys = [:cond, :susp, :skew, :kurt, :bind]
keys_tex_doc = ["\$\\Sigma\$", "\$\\chi\$", "\$S\$", "\$K\$",  "\$B\$"]
pred_tags = [:Y_P1, :Y_P2]
orig_tag = :Y_BS
display(kappa_list)
```

```julia
chosen_key = :kurt
chosen_keyword = kappa_list[3]
```

```julia
chosen_tag = :Y_P1

idx = findfirst(isequal(chosen_key), first_block_keys)
chosen_key_tex = keys_tex_doc[idx]

bc_dict, _ = RebekahMiriam.ComparisonRebekahMiriam.build_bhattacharyya_dicts_for_measurements(
    new_dict_cumulant_meas,
    first_block_keys,
    kappa_list,
    pred_tags,
    orig_tag,
    labels,
    trains
)

BCD = bc_dict[(chosen_key, chosen_tag, chosen_keyword)]

RebekahMiriam.HeatmapsRebekahMiriam.render_bhattacharyya_heatmap_for_measurements(
	BCD, 
	labels_int, 
	trains_int, 
	chosen_key, 
	chosen_tag, 
	chosen_keyword,
	overall_name,
	PROJECT_ROOT;
	key_tex=chosen_key_tex,
	save_file=save_flag
)

chosen_tag = :Y_P2

idx = findfirst(isequal(chosen_key), first_block_keys)
chosen_key_tex = keys_tex_doc[idx]

bc_dict, _ = RebekahMiriam.ComparisonRebekahMiriam.build_bhattacharyya_dicts_for_measurements(
    new_dict_cumulant_meas,
    first_block_keys,
    kappa_list,
    pred_tags,
    orig_tag,
    labels,
    trains
)

BCD = bc_dict[(chosen_key, chosen_tag, chosen_keyword)]

RebekahMiriam.HeatmapsRebekahMiriam.render_bhattacharyya_heatmap_for_measurements(
	BCD, 
	labels_int, 
	trains_int, 
	chosen_key, 
	chosen_tag, 
	chosen_keyword,
	overall_name,
	PROJECT_ROOT;
	key_tex=chosen_key_tex,
	save_file=save_flag
)
```

```julia
chk_dict, err_dict, ovl_dict =
    RebekahMiriam.ComparisonRebekahMiriam.build_overlap_error_and_ovl_dicts_for_measurements(
        new_dict_cumulant_meas,   # ext_dict
        first_block_keys,         # keys
        kappa_list,               # kappa_list (e.g., ["13570","13575",...])
        pred_tags,                # e.g., [:Y_P1, :Y_P2]
        orig_tag,                 # e.g., :Y_BS
        labels,                   # LBP axis (ratios)
        trains;                   # TRP axis (ratios)
        σ_floor = 1e-12
    )

for chosen_tag in (:Y_P1, :Y_P2)
    idx = findfirst(isequal(chosen_key), first_block_keys)
    chosen_key_tex = keys_tex_doc[idx]

    OVL = ovl_dict[(chosen_key, chosen_tag, chosen_keyword)]
    ERR = err_dict[(chosen_key, chosen_tag, chosen_keyword)]

    RebekahMiriam.HeatmapsRebekahMiriam.render_overlap_type_b_and_error_heatmaps_for_measurements(
        OVL,                 # ← Type-B (|μ_orig-μ_pred| / σ_orig)
        ERR,                 # → Error ratio
        labels_int,          # y-axis ticks
        trains_int,          # x-axis ticks
        chosen_key,
        chosen_tag,
        chosen_keyword,      # e.g., "13580" (κ token)
        overall_name,
        PROJECT_ROOT;
        key_tex = chosen_key_tex,
        save_file = save_flag,
        vmax_ovl = 3.0
    )
end
```
